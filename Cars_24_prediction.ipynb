```python
import pandas as pd
from sklearn.preprocessing import MinMaxScaler
numerical_features = df.select_dtypes(include=['number']).columns
scaler = MinMaxScaler()
df[numerical_features] = scaler.fit_transform(df[numerical_features])
print(df[numerical_features])
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from statsmodels.graphics.regressionplots import influence_plot
import statsmodels.formula.api as smf
import numpy as np
df.corr(numeric_only=True)
```

```python
import statsmodels.formula.api as smf
df = df.rename(columns={'No._of_Owners': 'No_of_Owners'})
model = smf.ols("On_Road_Price~Year+KM_Driven+Calculated_Score+No_of_Owners",data=df).fit()
model.summary()
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
plt.figure(figsize=(8, 6))
plt.scatter(df['On_Road_Price'], df['Calculated_Score'])
plt.xlabel('On_Road_Price')
plt.ylabel('Calculated_Score')
plt.title('Scatter Plot of On_Road_Price vs Calculated_Score')
plt.grid(True)
```

```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
df['Name'] = le.fit_transform(df['Name'])
df['Company'] = le.fit_transform(df['Company'])
df['Fuel_Type'] = le.fit_transform(df['Fuel_Type'])
df['City'] = le.fit_transform(df['City'])
df
```

```python
corr_matrix = df.corr()
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm')
plt.title('Correlation Heatmap')
plt.show()
```

```python
ml_km = smf.ols('KM_Driven~On_Road_Price',data = df).fit()
print(ml_km.tvalues, '\n', ml_km.pvalues)
(ml_km.rsquared,ml_km.rsquared_adj)
```

```python
ml_n = smf.ols('No_of_Owners~On_Road_Price',data = df).fit()
print(ml_n.tvalues, '\n', ml_n.pvalues)
(ml_n.rsquared,ml_n.rsquared_adj)
```

```python
ml_orp = smf.ols('On_Road_Price~Calculated_Score+KM_Driven+Fuel_Type+Company',data = df).fit()
print(ml_orp.tvalues, '\n', ml_orp.pvalues)
(ml_orp.rsquared,ml_orp.rsquared_adj)
```

```python
rsq_orp = smf.ols('On_Road_Price~Calculated_Score+KM_Driven+Fuel_Type+Company',data=df).fit().rsquared
vif_orp = 1/(1-rsq_orp)
print(vif_orp)
rsq = smf.ols('Calculated_Score~On_Road_Price',data=df).fit().rsquared
vif = 1/(1-rsq)
print(vif)
rsq = smf.ols('KM_Driven~On_Road_Price',data=df).fit().rsquared
vif = 1/(1-rsq)
print(vif)
rsq = smf.ols('Fuel_Type~On_Road_Price',data=df).fit().rsquared
vif = 1/(1-rsq)
print(vif)
rsq = smf.ols('Company~On_Road_Price',data=df).fit().rsquared
vif = 1/(1-rsq)
print(vif)
rsq = smf.ols('No_of_Owners~On_Road_Price',data=df).fit().rsquared
vif = 1/(1-rsq)
print(vif)
rsq = smf.ols('Year~On_Road_Price',data=df).fit().rsquared
vif = 1/(1-rsq)
print(vif)
rsq = smf.ols('Name~On_Road_Price',data=df).fit().rsquared
vif = 1/(1-rsq)
print(vif)
```

```python
def get_standardized_values(vals):
  return (vals-vals.mean())/vals.std()

plt.scatter(get_standardized_values(model.fittedvalues),
            get_standardized_values(model.resid))
plt.title('Residual Plot')
plt.xlabel('Standardized Fitted Values')
plt.ylabel('Standardized Residuals values')
plt.grid(True)
plt.show()
```

```python
import statsmodels.api as sm
qqplot=sm.qqplot(model.resid,line='q')
plt.title("Normal Q-Q plot of residuals")
plt.show()
```
